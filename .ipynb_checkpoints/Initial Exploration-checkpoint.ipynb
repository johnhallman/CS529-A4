```python
# takes a data directory and returns a dictionary mapping animal names to a list of images
def load_imgs(data_dir, full=False):
    imgs = {}
    dir_imgs = os.listdir(data_dir)
    if not full and len(dir_imgs) > 1000: dir_imgs = dir_imgs[:1000]
    for img_name in dir_imgs:
        id_num = int(img_name.split('_')[-1].split('.')[0])
        img = plt.imread(data_dir + img_name)
        imgs[id_num] = img
    return imgs

def load_json(json_file):
    json_data = open(json_file)
    data = json.load(json_data)
    json_data.close()
    return data
```

```python
import tensorflow as tf
from keras.models import Sequential, Model
from keras.layers import Dense, Conv2D, Flatten, MaxPooling2D, Activation, Input, Dropout, Add, concatenate
from keras.layers.normalization import BatchNormalization
from keras.callbacks import ModelCheckpoint
from keras.optimizers import Adam

#from keras.applications.resnet50 import ResNet50 # pretrained model, default size 224x224
#ResNet50(include_top=True, weights='imagenet', input_tensor=None, input_shape=None, pooling=None, classes=1000)

a1 = Input(shape=(2,))
a2 = Input(shape=(2,))
b1 = Dense(5, activation='relu')(a1)
b2 = Dense(5, activation='relu')(a2)
b = concatenate([b1, b2])
b = Dense(5, activation='relu')(b)
c = Dense(2, activation='sigmoid')(b)
model = Model(inputs=[a1, a2], outputs=[c])
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

x_train = [np.array([[0, 1], [1, 1], [1, 1], [0, 1]]), np.array([[0, 0], [1, 0], [0, 1], [1, 1]])]
y_train = [np.array([[0, 0], [1, 0], [1, 1], [0, 1]])]
model.fit(x_train, y_train, epochs=3, verbose=2)
```

```python
def load_imgs(directory, entire=False):
    k_arbitrary = 100
    dir_imgs = os.listdir(directory)
    if not entire and len(dir_imgs) > k_arbitrary: dir_imgs = dir_imgs[:k_arbitrary]
    data = {}
    for img_name in dir_imgs:
        img = plt.imread(os.path.join(directory, img_name))
        id_num = int(img_name.split('_')[-1].split('.')[0])
        data[id_num] = {'img':img}
    return data


def load_questions(directory, data):
    with open(directory) as file:
        questions = json.load(file)['questions']
        for q in questions:
            if q['image_id'] not in data: continue
            data[q['image_id']][q['question_id']] = {'q':q['question']}
            
            
def load_answers(directory, data, k=1000):
    with open(directory) as file:
        answers = json.load(file)['annotations']
        for a in answers:
            if a['image_id'] not in data or a['question_id'] not in data[a['image_id']]: 
                continue
            data[a['image_id']][a['question_id']]['a'] = a['multiple_choice_answer']
        a_list = [a['multiple_choice_answer'] for a in answers]
    a_set = set(a_list)
    w_to_i = {}
    i_to_w = {}
    for i, w in enumerate(a_set):
        w_to_i[w] = i
        i_to_w[i] = w
    score = len(a_set) * [0]
    for w in a_list:
        score[w_to_i[w]] += 1
    w_sorted = np.flip(np.argsort(score))

    top_ans = [i_to_w[i] for i in w_sorted[:k]]
    return top_ans
        

def load_glove(directory):
    glove = {}
    with open(directory) as file:
        for line in file:
            words = line.split(' ')
            word, vector = values[0], np.asarray(values[1:], dtype='float32')
            glove[word] = vector
    return glove


def load_answer_key(data, answer_key):
    scores = {}
    
    pass


    
def digest_data(data, glove, answer_key):
    def s_to_v(s):
        s_cut = s[-1] if s[-1] == '?' else s
        return np.array([glove[w.lower()] for w in s_cut.split(' ')])
    
    x, y = [[], []], []
    for _, v in data.items():
        for k, pair in v.items():
            if k == 'img': continue
            x[0].append(v['img'])
            x[1].append(s_to_v(pair['q']))
            y.append(answer_key[pair['a']])
    return x, y


def load_data(directory, entire=False):
    train_data = load_imgs(os.path.join(directory, "train_images"), entire=entire)
    val_data = load_imgs(os.path.join(directory, "val_images"), entire=entire)
    load_questions(os.path.join(directory, "train_questions.json"), train_data)
    load_questions(os.path.join(directory ,"val_questions.json"), val_data)
    top_test_anws = load_answers(os.path.join(directory, "train_annotations.json"), train_data)
    top_val_anws = load_answers(os.path.join(directory, "val_annotations.json"), val_data)
    glove = load_glove("glove.6B/glove.6B.300d.txt")
    answer_key = load_answer_key(train_data)
    x_train, y_train = digest_data(train_data, glove, answer_key)
    x_val, y_val = digest_data(val_data, glove, answer_key)
    return x_train, y_train, x_val, y_val
```
